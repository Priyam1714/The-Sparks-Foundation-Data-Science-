```python
# AIM: -To make a model that predicts marks by the number of hours a student have studied.(by using linear regression)
```

```python
# Step 1: Importing all the necessary libraries needed to design this model
from sys import displayhook
import numpy as np
from numpy.core.fromnumeric import reshape
import pandas as pd
import matplotlib.pyplot as plt
from pandas.io.formats import style
```

```python
#Step 2: Accessing the Raw dataset with the help of Pandas Library

url="https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv"
student_data=pd.read_csv(url)
print("Data is loaded")
print(student_data.head(10))
```

```python
#Step 3: Loading the raw accessed data and fiting it into a plot named(Study Hours V/S Obtained marks)
#Legend 
# 1.X-axis: Nunmber of Hours the student have studied.
# 2.Y-axis: The marks the student got.


student_data.plot(x="Hours",y="Scores", style="*")
plt.title("Study Hours V/S Obtained Marks")
plt.xlabel("Hours")
plt.ylabel("Marks Obtained")
plt.show()
```

```python

x = student_data.iloc[:, :-1].values  
y = student_data.iloc[:, 1].values  
```

```python
#Step 4: From Sk learns library, Loading the train,test,split to convert our raw data into data that will be used to train the model.

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#Step 5: Loading the Linear Regression from sk learn and loading the training data obtained from train_test_split into our Linear Regression model.

from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
print("Training Complete")
```

```python
#Step 6: Plotting a linear line, by using our obtained data from previous block by fitting it into the formula. (y=m*x+c)
#Legend: x=input, m=reg.coef_ , y=intercept,(m=Slope if said in simple words)

line=reg.coef_*x+reg.intercept_
plt.scatter(x,y)
plt.plot(x,line)
plt.show()
```

```python
#Step 7: Now printing our x_test data obtained from the Linear Regression model we trained earlier and giving that as input to get the y_predicted

print(x_test)
y_pred=reg.predict(x_test)
```

```python

df=pd.DataFrame({'Actual Data':y_test,'Predicted Data':y_pred})
print (df)
```

```python
# Step 8: Giving input as 9.5 hours into our trained model to get the final result(93,691...). This block of code can be modified to take custom input from the user.
#Note- reshaping of the float data h is necessary in order to fit the data in the model.

 
h=9.25
hours=np.reshape(h,(-1,1))
#hours=input("Enter your own data to predict")
own_pred=reg.predict(hours)
print("Student studied for {} number of hours".format(hours))
print("and scored {}".format(own_pred[0]))
```
